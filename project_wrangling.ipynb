## 3- Data Cleaning


### 3-1 Fixing Quality Issues


```python
#copy from the dataframe
df_clean = df.copy()
df_clean.head()
```

#### A- Define :
-  Replace (?) from column ['normalized_losses'] with the mean column

#### B- Code :

```python
df['normalized_losses'] = df['normalized_losses'].replace('?', np.nan).astype(float)
```

```python
mean_normalized_losses = df['normalized_losses'].mean()
mean_normalized_losses
```

```python
df['normalized_losses'].fillna(mean_normalized_losses, inplace=True)
```

#### C-Test:

```python
df.head(3)
```

#### A- Define :
- Replace (?) from column ['price'] with the mean column

#### B- Code :

```python
df['price'] = df['price'].replace('?', np.nan).astype(float)
```

```python
mean_price = df['price'].mean()
mean_normalized_losses
```

```python
df['price'].fillna(mean_price, inplace=True)
```

#### C-Test:

```python
df['price'].value_counts()
```

#### A- Define :
- Remove blank rows (?) from columns ['bore', 'stroke', 'num_of_doors']

#### B- Code :

```python
df[['bore', 'stroke', 'num_of_doors', 'horsepower']] = df[['bore', 'stroke', 'num_of_doors', 'horsepower']].replace('?', np.nan)
```

```python
df.dropna(subset=['bore', 'stroke', 'num_of_doors', 'horsepower'], inplace=True)
```

#### C-Test :

```python
df.shape
```

```python
df
```

#### A-Define :
- ['normalized_losses','horsepower', 'peak_rpm', 'price' ] Transfer to int

#### B- Code :

```python
df['normalized_losses'] = df['normalized_losses'].astype('int64')
df['horsepower'] = df['horsepower'].astype('int64')
df['peak_rpm'] = df['peak_rpm'].astype('int64')
df['price'] = df['price'].astype('int64')
```

#### C-Test :

```python
df.info()
```

#### A-Define :
- ['bore','stroke'] Transfer to float

#### B- Code :

```python
df['bore'] = df['bore'].astype('float64')
df['stroke'] = df['stroke'].astype('float64')
```

#### C- Test :

```python
df.info()
```

#### A- Dfine :
- Replace( ' - ' ) from column[ 'make' ] to space

#### B- Code :

```python
df['make'] = df['make'].str.replace('-', ' ')
```

#### C- Test :

```python
df.head()
```

#### A- Define :
- Convert text from column [ 'num_of_doors' ] to int => from (four , two) to (4,2)

#### B- code :

```python
df['num_of_doors'] = df['num_of_doors'].replace({'two':2, 'four':4}) 
```

#### C- Test :

```python
df['num_of_doors'].value_counts()
```

## 5- Data Visualization


```python
# Look at the data
df.hist(figsize=(13,13));
```

```python
plt.figure(figsize=[20, 7.5])
df.groupby('make')['price'].mean().plot(kind='bar', alpha=0.7, color='blue', label='Average Price')
plt.title('Average Car Price by Make', size=20)
plt.xlabel('Car Make', size=15)
plt.ylabel('Average Price', size=17)
plt.legend()
plt.show();
```

```python
plt.figure(figsize=[5, 5])
df.groupby('fuel_type')['price'].mean().plot(
            kind='pie', autopct='%1.1f%%', colors=['blue', 'orange'], startangle=90
           )
plt.title('Average Car Price by Fuel Type', size=10)
plt.ylabel('')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=[5, 5])
df.groupby('num_of_doors')['price'].mean().plot(
            kind='pie', autopct='%1.1f%%', colors=['blue', 'orange'], startangle=90
           )
plt.title('Average Car Price by num doors', size=10)
plt.ylabel('')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[20, 7.5])
df.groupby('body_style')['price'].mean().plot(kind='bar', alpha=0.7, color='blue', label='Average Price')
plt.title('Average Car Price by body style', size=20)
plt.xlabel('body style', size=15)
plt.ylabel('Average Price', size=17)
plt.legend()
plt.show();
```

```python

plt.figure(figsize=[20, 7.5])
df.groupby('body_style')['horsepower'].mean().plot(kind='bar', alpha=0.7, color='blue', label='Average Price')
plt.title('Average Car Price by horsepower ', size=20)
plt.xlabel('body style', size=15)
plt.ylabel('Average horsepower', size=17)
plt.legend()
plt.show();
```

```python

plt.figure(figsize=[20, 7.5])
df.groupby('body_style')['normalized_losses'].mean().plot(kind='bar', alpha=0.7, color='blue', label='Average Price')
plt.title('Average Car Price by body style', size=20)
plt.xlabel('body style', size=15)
plt.ylabel('Average losses', size=17)
plt.legend()
plt.show();
```
